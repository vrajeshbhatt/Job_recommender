```python
import numpy as np
import pandas as pd
```

```python
# loading dataset
data = pd.read_csv('D:/Job_prediction/job_desc.csv')
data
```

```python
# values of description
desc = data.iloc[:,-1]
```

```python
desc[0]
```

```python
import spacy
#Initializing the spacy engine and then use it to process the text:
nlp = spacy.load('en_core_web_sm')

# finding noun phrases and appendin to a list
nouns = []
for i in desc:
    doc = nlp(i)
    for chunk in doc.noun_chunks:
        nouns.append(chunk)
```

```python
# converting each value as str
text_nouns = [str(x) for x in nouns]
text_nouns
```

```python
#making a data frame
df_nouns = pd.DataFrame(text_nouns, columns = ['description'])
```

```python
df_nouns['description']
```

```python
from nltk.corpus import stopwords
```

```python
test = pd.DataFrame()
## Lower case
df_nouns['description'] = df_nouns['description'].apply(lambda x: " ".join(x.lower()for x in x.split()))
## remove tabulation and punctuation
df_nouns['description'] = df_nouns['description'].str.replace('[^\w\s]',' ')
## digits
df_nouns['description'] = df_nouns['description'].str.replace('\d+', '')

#remove stop words
stop = stopwords.words('english')
df_nouns['description'] = df_nouns['description'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```
